## **2. Pra-Pemrosesan Data**

### Data Training & Data Testing

```python
df["quality"].unique()
```

 proses pra-pemrosesan data khususnya untuk kolom "quality" yang menjadi target atau label dalam dataset. Perintah df["quality"].unique() digunakan untuk melihat nilai-nilai unik dalam kolom "quality", yang menunjukkan bahwa terdapat enam kelas berbeda yaitu 3, 4, 5, 6, 7, dan 8. Ini mengindikasikan bahwa kualitas wine dalam dataset dinilai dalam rentang tersebut, dengan nilai yang lebih tinggi menunjukkan kualitas yang lebih baik.

```python
print(df.quality.value_counts())
```

hasil dari df.quality.value_counts() menampilkan jumlah data untuk masing-masing nilai kualitas. Diketahui bahwa kelas kualitas 5 dan 6 mendominasi dataset dengan jumlah masing-masing 362 dan 341. Sedangkan kualitas 3, 4, 7, dan 8 memiliki jumlah data yang jauh lebih sedikit, terutama kelas 3 dan 8 yang hanya terdiri dari masing-masing 6 dan 13.

```python
df["quality"].value_counts(normalize = True)

scaler = MinMaxScaler()

min_max_df = scaler.fit_transform(df.drop("quality", axis = 1))
min_max_df = pd.DataFrame(min_max_df, columns = df.columns[:-1])
X = min_max_df
y = df['quality']

oversample = SMOTE()

X, y = oversample.fit_resample(X, y)
X.shape, y.shape
y.value_counts()
```

proses normalisasi data dan penyeimbangan kelas menggunakan teknik SMOTE (Synthetic Minority Over-sampling Technique) sebelum melakukan pemisahan data latih dan data uji. Data selain kolom target (quality) dinormalisasi menggunakan MinMaxScaler agar semua fitur berada dalam rentang 0 hingga 1, sehingga model tidak berat sebelah terhadap fitur tertentu karena perbedaan skala. Kemudian, dilakukan pemisahan variabel independen (X) dan dependen (y) dengan X berisi semua fitur dan y adalah label kualitas wine.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

data dibagi menjadi data latih dan uji  menggunakan train_test_split, memastikan distribusi data yang adil untuk pelatihan dan evaluasi model.

## **3. Pemrosesan Data**

```python
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=4),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    "GradBoost": GradientBoostingClassifier(),
    'MLP Neural Network': MLPClassifier(max_iter=1000),
}

evaluation_results = []

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # metrik evaluasi
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # hasil
    evaluation_results.append({
        'Model': model_name,
        'Accuracy': round(accuracy, 4),
        'Precision': round(precision, 4),
        'Recall': round(recall, 4),
        'F1 Score': round(f1, 4)
    })
```

dilakukan evaluasi terhadap beberapa algoritma klasifikasi, yaitu Decision Tree, K-Nearest Neighbors (KNN), Support Vector Machine (SVM), Gradient Boosting, dan MLP Neural Network. Masing-masing model dilatih menggunakan data latih (X_train dan y_train) dan diuji menggunakan data uji (X_test). Proses evaluasi dilakukan dengan menghitung empat metrik performa utama, yaitu akurasi, presisi, recall, dan F1-score dengan pendekatan rata-rata berbobot (average='weighted'). Hasil evaluasi dari tiap model disimpan dalam bentuk list of dictionary, yang berisi nama model dan nilai metrik evaluasinya.

## **4. Evaluasi Model**

### Matrik Evaluasi Model

```python
# DataFrame
results_df = pd.DataFrame(evaluation_results)

fig = go.Figure(data=[go.Table(
    header=dict(
        values=[f"<b>{col}</b>" for col in results_df.columns],
        fill_color='white',
        line_color='black',
        align='center',
        font=dict(color='black', size=12)
    ),
    cells=dict(
        values=[results_df[col] for col in results_df.columns],
        fill_color='white',
        line_color='black',
        align='center',
        font=dict(color='black', size=12)
    )
)])

fig.update_layout(
    title=dict(
        text="Perbandingan Evaluasi Model",
        x=0.5,
        xanchor='center',
        font=dict(color='black', size=20)
    ),
    margin=dict(t=30)
)

fig.show()
```

 hasil visualisasi evaluasi performa lima model klasifikasi dalam bentuk tabel. Metrik yang ditampilkan mencakup akurasi, presisi, recall, dan F1-score. Berdasarkan tabel, model Gradient Boosting menunjukkan performa terbaik dengan nilai akurasi 0.8103 dan F1-score tertinggi sebesar 0.8056, yang mengindikasikan bahwa model ini paling konsisten dalam mengklasifikasikan data dengan benar. Sebaliknya, model SVM memiliki performa terendah di semua metrik, menunjukkan bahwa model ini kurang sesuai untuk dataset yang digunakan.

```python
classifier = GradientBoostingClassifier()
classifier.fit(X_train, y_train)
```

proses pelatihan model Gradient Boosting Classifier, Model diinisialisasi menggunakan GradientBoostingClassifier() dari pustaka sklearn.ensemble dan kemudian dilatih dengan data pelatihan X_train dan y_train melalui perintah .fit().

```python
y_pred=classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.to_numpy().reshape(len(y_test),1)),1))
final_classifier = classifier
```

proses prediksi dan pencocokan hasil, Model yang telah dilatih digunakan untuk memprediksi label dari data uji X_test dengan menggunakan metode .predict(). Hasil prediksi y_pred kemudian dibandingkan dengan label aktual y_test. Hasil dari keduanya ditampilkan berdampingan dengan menggunakan np.concatenate, yang membantu dalam memverifikasi apakah prediksi yang dilakukan sudah sesuai dengan label sebenarnya.
